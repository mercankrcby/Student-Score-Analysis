```python
#Standard Deviation and Mean Calculation -- READING SCORE 
readingScore_mean = df['reading score'].mean()
print("Mean:",readingScore_mean)
readingScore_std = df['reading score'].std()
readingScore_std
print("Standard Deviation:",readingScore_std)
```

```python
#Standard Deviation and Mean Calculation -- MATH SCORE 
mathScore_mean = df['math score'].mean()
print("Mean:",mathScore_mean)
mathScore_std = df['reading score'].std()
mathScore_std
print("Standard Deviation:",mathScore_std)
```

```python
#Standard Deviation and Mean Calculation -- WRITING SCORE 
writingScore_mean = df['writing score'].mean()
print("Mean:",writingScore_mean)
writingScore_std = df['writing score'].std()
writingScore_std
print("Standard Deviation:",writingScore_std)
```

```python
# Create a list of colors (from iWantHue)
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]

# Create a pie chart
plt.pie(
    # using data total)arrests
    df['Count'],
    # with the labels being officer names
    labels=df['Gender'],
    # with no shadows
    shadow=False,
    # with colors
    colors=colors
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.tight_layout()
plt.show()
```
